```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df = df.drop('User ID',axis=1)
```

```python
df.head()
```

```python
df = pd.get_dummies(df,drop_first=True)
```

```python
df
```

```python
y = df[['Purchased']]
x = df.drop('Purchased',axis=1)
```

```python
x.head()
```

```python
y.head()
```

```python
# test-size - 20%
# random_state = 70
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=70)
```

# KNN

```python
KNN  = KNeighborsClassifier(n_neighbors=3)
KNN.fit(x_train,y_train)
y_pred  = KNN.predict(x_test)
```

```python
y_pred
```

```python
prediction = pd.DataFrame({
    'actul' : y_test.squeeze(),
    'Pridicted' : y_pred.squeeze()
})
prediction
```

```python
# age :25 
# salary:30000
# gender : False
y_pred_data  = KNN.predict([[25,30000,False]])
y_pred_data

if y_pred_data == 1:
    print('Purchased')
else:
    print('Not Purchased')
    
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
conf_matrix = confusion_matrix(y_test,y_pred)
conf_matrix
```

```python
sns.heatmap(conf_matrix,annot=True,
            linecolor='black',
            linewidths=2,
           )
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
TN = conf_matrix[0,0]
TP = conf_matrix[1,1]
FP = conf_matrix[0,1]
FN = conf_matrix[1,0]

TN,TP,FP,FN
```

```python
accuracy = (TP+TN)/np.sum(conf_matrix)
precision = TP/(TP+FP)
recall = TP/(TP+FN)
print("Accuracy : ",accuracy)
print("Precisionprecision : ",precision)
print("Recallrecall : ",recall)
```
